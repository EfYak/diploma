```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_tsdata(filename):
    '''
    load time series data from binary file
    return time,z,y,x (i.e. dim = 4 x nTimeStamps); time in ns
    '''
    data = np.fromfile(filename, dtype=np.dtype('f'))
    ##normdata=(np.array(data, dtype=float)-127)/128
    lsize = int(len(data)/4)
    normdata = np.reshape(data, (lsize, 4))
    ##normdata.dtype=np.complex
    return normdata.T
```

```python
myfn = r'D:\Efim\thesis\vibration-master\sampledata\dat0-05.bin'
yy = read_tsdata(myfn)
```

```python
plt.figure(figsize=(20,7))
print (yy.shape)
plt.plot(yy[0,:]/1e9, yy[2,:])
plt.xlabel('Time (s) since start')
plt.ylabel('Raw g acceleration (Z direction)')

plt.show()
```
